```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_excel("data/score.xlsx",index_col='지원번호')
```

```python
# int 타입에 cm문자 추강방법 
#df['키'] = df['키'].astype(str) + 'cm'

################### 함수사용해서 추가 ###################
# def add_cm(height):
#     return str(height)+'cm'
# df['키'] =df['키'].apply(add_cm)

df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['평균'] = df['합계'] / 5
df['등급'] = 'F'


#함수를 사용해서 90점이상은 합격 ,90점미만은 불합격 
def rank(score):
    result =''
    if score >90:
     result ="합격"
    else: 
        result = "불합격"
    return result

df['등급'] = df['평균'].apply(rank)
```

```python
### 90점이상  A  ,80점 이상 B 70점이상c, 60점이상D 60점미만 F
def rank(score):
    result =''
    if score >=90:result ="A"
    elif score >=80:result = "B"
    elif score >=70:result = "C"
    elif score >=60:result = "D"
    elif score <60:result = "F"  
    return result

#df['등급'] = df['평균'].apply(rank)
```
