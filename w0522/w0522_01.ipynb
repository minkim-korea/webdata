```python
import pandas as pd
import numpy as np
```

```python
data = {
    '이름':['홍길동','유관순','이순신','강감찬','김구','김유신','홍길자','홍길순'],
    '원':['19,700','18,400','16,800','18,700','18,800','20,200','18,800','19,000'],
    '우대':[90,40,80,40,15,80,55,100]
}
```

```python
df =pd.DataFrame(data)
df['최종'] = 0
df['원'] = df['원'].str.replace(",","").astype(int)
```

```python
## 1. 최종 컬럼 생성, 원 + 원*우대*0.01 = 최종컬럼에 추가하시오.
# 모든 최종 컬럼에 원 + 원*우대*0.01 적용

# df['원'] =  df['원'].str.replace(",","").astype(int)
# df
# df['최종'] = 0
# df['최종'] = df['원'] + df['원']*df['우대']*0.01
```

```python
### 우대 50이상 : 원 + 원*우대*0.01, 우대 50미만 : 원 부분만 추가하시오.
filt = df['우대']>=50
df.loc[filt,'최종'] = df['원'] + df['원']*df['우대']*0.01
df.loc[~filt,'최종'] = df['원']
df
```

```python
###함수를 사용해서 진행 ,if문 비교 각각 데이터 전달해서 진행  
def change(data):
    if data['우대']>=50 : return data['원'] +data['원']*data['우대']*0.01
    else: return data['원'] 

df['최종'] = df.apply(change,axis=1) # axis: row 행 데이터를 모두 전달 
df
```
