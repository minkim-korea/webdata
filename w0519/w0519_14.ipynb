```python
import pandas as pd 
 
df = pd.read_csv('data/stuscore (1).csv',index_col='번호')
df
```

```python
### startswith 함수 - 해당글자로 시작되는 문자검색 
filt = df['이름'].str.startswith('홍')
df[filt]
```

```python
#### endswith -해당글자로 끝나는 문자 검색 
df[df['이름'].str.endswith("신")]
df[df['이름'].str.endswith("s")]
```

```python
####contains - 해당글자가 포함되어 있는 문자검색 
df[df['이름'].str.contains("s")]
df[df['이름'].str.contains("순")] 
df[~df['이름'].str.contains("순")] # ~ 은 not 을 의미 
```

```python
### a 가들어가 있는 문자를 검색하시오 . -40개 
df[df['이름'].str.contains('a')].count() #-40개
df[df['이름'].str.contains('A')]   # -8개
df[df['이름'].str.contains('A') | df['이름'].str.contains('a')] # 48개
```

소문자로 변경후 글자가 있는지 검색 

```python
df[df['이름'].str.lower().str.contains("a")].count()
df['이름'].str.lower()# 모든 글자 소문자로 취환

#모든글자를 소문자로 변경후 a가포함되어 있는 문자 검색 
df[df['이름'].str.lower().str.contains('a')].count()

#모든글자 대문자로 변경 
df['이름'].str.upper()
df[df['이름'].str.upper().str.contains('A')].count()
```

```python
dfa =pd.read_excel("data/score.xlsx",index_col="지원번호")
#홍이 포함되어있는 학생의 국영수과 성적을 출력하시오.
#column기준으로 출력
dfa[dfa['이름'].str.contains("홍")][['이름',"국어","영어","수학","과학"]]
# row를 기준으로출력
dfa.loc[dfa['이름'].str.contains("홍"),['이름',"국어","영어","수학","과학"]]
dfa.loc[dfa['이름'].str.contains("홍"),'국어':'과학']
#sw특기가 python java 글자가 포함되어 있는 학생의 이름 ,학교 키 sw특기를 출력하시오 
dfa[dfa['SW특기'].str.lower().str.contains("python") | dfa['SW특기'].str.lower().str.contains("java")]



### 3. 리스트 형태로 검색

langs = ['python','java']

##글자가 일치하는 문자검색 
# dfa[dfa['SW특기'].str.lower().str.contains(langs)] 에러
dfa[dfa['SW특기'].str.lower().isin(langs)]
```
