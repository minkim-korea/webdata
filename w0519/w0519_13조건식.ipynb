```python
# 조건식 &연산자
df[df['키']>=185 & (df['수학']>=80)]
df[~((df['키']>=185) & (df['수학']>=80))]
```

```python
# 키가 185이하이면서  또는  수학이 70점이상
df[((df['키']<=185) | (df['수학']>=70))]
df[~((df['키']<=185) | (df['수학']>=70))]
```
