```python
import pandas as pd 
df = pd.read_excel('file/score.xlsx',index_col='지원번호')
df.head(2)
```

### 데이터 출력 

```python
#1. rows 로 출력 
df.head()
df.tail()
df.head(2)
# 2번째 :rows 출력 -loc 형태는 index값으로 출력하는거임 
df.loc[['1번','3번']] # index 1번,3번 출력 
df.loc[:] #전체출력은 가능 
df.loc['3번':] # 3번부터 전체출력 
df.loc['1번':'3번'] #1번부터 3번까지 
df.loc['2번':'8번'] #2번부터 8번까지 

#3번째 rows출력 - iloc : 번호로 출력 0부터시작 
df.iloc[0] #0번째 있는 row출력 
df.iloc[3] #3번째 있는 row출력 

df.iloc[:] # 전체출력 
df.iloc[3:] # 3:끝까지 



```

```python
### 컬럼 출력  [] 안에 컬럼을 넣으면됨 
df['이름']
df[['이름','학교','SW특기']]
#컬럼명  모두 가져오기  -리스트형태 
df.columns[0]
df.columns[1]
df.columns

#인덱스 출력
df.index

df['이름']        
df.columns[0]     # 이름 
df[df.columns[0]] # df['이름'] 과같음 

## 전체출력 -컬럼별로 출력  
for name in df.columns:
    print(df[name])
```
